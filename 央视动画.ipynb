```python
import urllib
import requests
import re
```

```python
result={}
hot_index={}
result_sort={}
for page in range(1,46+1):
    url='http://api.cntv.cn/List/getVideoDocumentary?id=CHAL1460955899450127&serviceId=tvcctv&n=20&p='+str(page)+'&area=中国大陆&cb=Callback'
    data=requests.get(url)
    content=data.content.decode('utf-8')
    pattern_name=re.compile(r"(?<=\"name\":\").+?(?=\")")
    m_name=pattern_name.findall(content)
    pattern_year=re.compile(r"(?<=\"year\":\").+?(?=\D)")
    m_year=pattern_year.findall(content)
    for i in range(len(m_year)):
        if m_year[i] in result.keys():
            result[m_year[i]].append(m_name[i])
        else:
            result[m_year[i]]=[m_name[i]]
```

```python
for key in range(1949,2019):
    if str(key) in result.keys():
        hot_index[str(key)]=[]
        for name in result[str(key)]:
            name_zh=re.search(u"([\u4e00-\u9fa5]|\d|[a-z]|[A-Z])+",name).group()
            url2='http://zhishu.sogou.com/index/searchHeat?kwdNamesStr='+name_zh+'&timePeriodType=YEAR&dataType=SEARCH_ALL&queryType=INPUT'
            index_text=requests.get(url2).content.decode('utf-8')
            index=re.search(r'(?<=\"avgPv\":).+?(?=\D)',index_text)
            if index==None:
                hot_index[str(key)].append(0)
            else:
                index=int(index.group())
                hot_index[str(key)].append(index)
```

```python
for key in range(1949,2019):
    if str(key) in hot_index.keys():
        hot_sort=sorted(hot_index[str(key)],reverse=True)
        result_sort[str(key)]=[]
        for sort_index in hot_sort:
            index1=hot_index[str(key)].index(sort_index)
            result_sort[str(key)].append(result[str(key)][index1])
            hot_index[str(key)][index1]=-1
```

```python
for key in range(1949,2019):
    if str(key) in hot_index.keys():
        print(key,result_sort[str(key)])
```
